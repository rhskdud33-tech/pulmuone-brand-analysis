```python
오뚜기메타 = own_oto_meta.copy()
오뚜기메타.head(5)
```

```python
# 열 이름 수정
오뚜기메타 = 오뚜기메타.rename(columns={'브랜드': '브랜드1'})
오뚜기메타 = 오뚜기메타.rename(columns={'브랜드라인': '브랜드2'})
오뚜기메타 = 오뚜기메타.rename(columns={'별점': '평균 별점'})
오뚜기메타 = 오뚜기메타.rename(columns={'리뷰수': '리뷰 수'})
오뚜기메타 = 오뚜기메타.rename(columns={'배송비': '배송'})
오뚜기메타.head(1)
```

```python
# 비비고메타 리뷰 평점 열 삭제
오뚜기메타 = 오뚜기메타.drop(columns=['카테고리', '상품_url'])

# 순위 열 생성
오뚜기메타['순위'] = range(1, len(오뚜기메타) + 1)
오뚜기메타.head(1)
```

```python
오뚜기메타['판매처'] = '오뚜기자사몰'
오뚜기메타.head(1)
```

```python
# 할인율 열 생성 + 값 추가 (할인율 = (정가 - 할인가) / 정가 * 100). 뒤에 소수점 첫째 자리에서 반올림. 소수점 안 나오게 하기
오뚜기메타['할인율'] = (((오뚜기메타['정가'] - 오뚜기메타['할인가']) / 오뚜기메타['정가']) * 100).round().astype(int)
오뚜기메타.head(1)
```

```python
오뚜기메타['만두상세내용'] = None
오뚜기메타['대체지표'] = None
```

```python
오뚜기메타.columns
```

```python
오뚜기메타 = 오뚜기메타[
    ['브랜드1', '브랜드2', '판매처', '순위', '상품ID', '상품명',
     '평균 별점', '리뷰 수', '정가', '할인가', '할인율',
     '만두상세내용', '대체지표', '배송']
]

오뚜기메타.head()
```

```python
네이버메타 = pd.read_csv('메타데이터_네이버.csv')
쿠팡메타 = pd.read_csv('메타데이터_쿠팡.csv')
```

```python
# 네이버메타 브랜드2 열에 판매처 열의 값 복사
네이버메타['브랜드2'] = 네이버메타['판매처']

# 네이버메타 판매처에 '네이버'입력
네이버메타['판매처'] = '네이버'
네이버메타.head(1)
```

```python
# 네이버메타 브랜드2의 모든 행 출력
네이버메타['브랜드2']
```

```python
# 풀무원메타, 비비고메타, 오뚜기메타, 네이버메타, 쿠팡메타 병합. 이름은 'total_meta'로. 브랜드1, 브랜드2, 판매처, 순위, 상품ID, 상품명, 평균 별점, 리뷰 수, 정가, 할인가, 할인율, 만두상세내용, 대체지표, 배송 열로 구성
total_meta = pd.concat([풀무원메타, 비비고메타, 오뚜기메타, 네이버메타, 쿠팡메타], ignore_index=True)
total_meta.head()
```

```python
# total_meta 저장
total_meta.to_csv('total_meta.csv', index=False)
```
